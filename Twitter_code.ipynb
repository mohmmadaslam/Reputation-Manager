```python
def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(score)))
```

```python
tweet_text['compound']=None
analyzer = SentimentIntensityAnalyzer()
score = analyzer.polarity_scores('')
```

```python
for i in range(len(tweet_text.index)):
    score = analyzer.polarity_scores(tweet_text['tweet'][i])
    tweet_text.iloc[i:i+1,6:7] = score['compound']
```

```python
tweet_text
```

```python
tweet_text['label'] = 0
tweet_text.loc[tweet_text['compound'] > 0.2, 'label'] = 1
tweet_text.loc[tweet_text['compound'] < -0.2, 'label'] = -1
tweet_text
```

```python
drop_label=tweet_text.index[tweet_text['label'] == 0].tolist()
tweet_text=tweet_text.drop(tweet_text.index[drop_label])
tweet_text.reset_index(drop=True,inplace=True)
tweet_text
```

```python
tweet_text['norm_follower'] = 0
tweet_text['norm_follower'] = tweet_text['follower_count']
tweet_text['norm_follower'] =(tweet_text['norm_follower']-tweet_text['norm_follower'].min())/(tweet_text['norm_follower'].max()-tweet_text['norm_follower'].min())
```

```python
tweet_text['weighted'] = tweet_text['norm_follower']*tweet_text['compound']
tweet_text.sort_values(by=['weighted'])
```
